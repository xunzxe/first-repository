```python
from random import *

class Charater:

    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def __str__(self):
        return 'ì´ë¦„ : {0} ë ˆë²¨ : {1} HP : {2} ê³µê²©ë ¥ : {3} ë°©ì–´ë ¥ : {4}'.format(self.name, self.level, self.hp, self.attack, self.defense)

    def is_alive(self):
        if self.hp > 0: # ì²´ë ¥ì´ 0ë³´ë‹¤ ë§ìœ¼ë©´ True ì ìœ¼ë©´ False / True ë©´ ìƒì¡´ False ì‚¬ë§
            return True
        else:
            return False


    def take_damage(self, damage):
        if damage - self.defense > 0: # ë°›ì€ ë°ë¯¸ì§€ê°€ í”Œë ˆì´ì–´ ë°©ì–´ë ¥ ë³´ë‹¤ ë§ìœ¼ë©´ ë°ë¯¸ì§€ë¥¼ ì…ëŠ”ë‹¤
            print('{0} ì—ê²Œ {1} ì˜ ë°ë¯¸ì§€ê°€ ë“¤ì–´ì™”ìŠµë‹ˆë‹¤.'.format(self.name, damage))
            print('ë°©ì–´ë ¥ {0}ë§Œí¼ ë°ë¯¸ì§€ë¥¼ ëœ ë°›ìŠµë‹ˆë‹¤.'.format(self.defense))
            self.hp -= (damage - self.defense)
            print('ë°›ì€ ë°ë¯¸ì§€ëŠ” {0} ì…ë‹ˆë‹¤. í˜„ì¬ {1} ì²´ë ¥ì€ {2} ì…ë‹ˆë‹¤.'.format((damage - self.defense), self.name, self.hp))

        else: # ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ ë°ë¯¸ì§€ë¥¼ ë°›ì§€ ì•ŠëŠ”ë‹¤.
            print('{0} ì—ê²Œ {1} ì˜ ë°ë¯¸ì§€ê°€ ë“¤ì–´ì™”ìŠµë‹ˆë‹¤.'.format(self.name, damage))
            print('ë°©ì–´ë ¥ {0}ë§Œí¼ ë°ë¯¸ì§€ë¥¼ ëœ ë°›ìŠµë‹ˆë‹¤.'.format(self.defense))
            print('ë°©ì–´ë ¥ì´ ë°›ì€ ë°ë¯¸ì§€ ë³´ë‹¤ ë†’ê¸° ë•Œë¬¸ì— ì²´ë ¥ì´ ê°ì†Œí•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.')


    def attack_target(self, enemy): # ëª¬ìŠ¤í„°ë¥¼ ê³µê²©í•  ë–„ 1 ~ í”Œë ˆì´ì–´ ê³µê²©ë ¥ ì‚¬ì´ì˜ ëœë¤í•œ ìˆ«ìë¡œ ê³µê²©í•œë‹¤.
        print('{0}ì´(ê°€) {1}ì„(ë¥¼) ê³µê²©í•©ë‹ˆë‹¤.'.format(self.name, enemy.name))
        attackDamage = randint(1, self.attack+1)
        print('{0}ì˜ ê³µê²©ì„ ê°€í•©ë‹ˆë‹¤.'.format(attackDamage))
        enemy.take_damage(attackDamage)

class Player(Charater):

    def __init__(self, name): # Charater ë¥¼ ìƒì† ë°›ìŒ
        # self.name = input('í”Œë ˆì´ì–´ ì´ë¦„ì„ ì§€ì •í•´ì£¼ì„¸ìš”. : ')
        self.name = name
        self.level = 1
        self.hp = 100
        self.attack = 25
        self.defense = 5
        self.exp = 0
        super().__init__(self.name, self.level, self.hp, self.attack, self.defense) # Playerì— ì§€ì •ëœ ê°’ì„ ë„£ì–´ ì¤Œ

    def __str__(self):
        return super().__str__() + ' ê²½í—˜ì¹˜ : {0}'.format(self.exp) # Charater í´ë˜ìŠ¤ì— ìˆëŠ” í”Œë ˆì´ì–´ ì •ë³´ì— ê²½í—˜ì¹˜ë¥¼ ì¶”ê°€

    def gain_experience(self, gainExp):

        self.exp += gainExp

        while True:
            if self.exp >= 50: # ê²½í—˜ì¹˜ê°€ 50 ì´ìƒì´ë¼ë©´
                self.level_up() # ë ˆë²¨ì—…ì„ í•˜ê³ 
                print('ë ˆë²¨ì´ ì˜¬ëìŠµë‹ˆë‹¤!')
                if self.exp - 50 != 0: # ë‚¨ì€ ê²½í—˜ì¹˜ë¥¼ ë”í•´ì£¼ëŠ” ì¡°ê±´
                    self.exp -= 50
                else:
                    self.exp = 0
            else:
                break

    def level_up(self):
        self.level += 1
        self.attack += 10
        self.defense += 5

    def __enter__(self):
       print(f'í™˜ì˜í•©ë‹ˆë‹¤, {self.name}!')
       return self

    def __exit__(self, a, b, c):
       print(f'{self.name}, í”Œë ˆì´ í•´ì£¼ì…”ì„œ ê°ì‚¬í•©ë‹ˆë‹¤.')

class Monster(Charater):

    def __init__(self, name, level):

        self.hp = randint(1, 3) * level
        self.attack = randint(1, 3) * level
        self.defense = randint(1, 3) * level
        self.exp = 6 * level
        super().__init__(name, level, self.hp, self.attack, self.defense)

    def __str__(self):
       return super().__str__() + ' ê²½í—˜ì¹˜ : {0}'.format(self.exp)
```
